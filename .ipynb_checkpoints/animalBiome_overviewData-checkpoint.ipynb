These scripts are written for AnimalBiome (animalbiome.com).

This notebook was used to generate graphics that describe attributes of the overall data set, such as cat breeds. Data is from customer surveys, and is metadata that accompanies the sequencing-based micribiome assessments AnimalBiome provides.

```python
#import libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import data, check it out
df = pd.read_table('path/file.txt')
df2 = df.drop_duplicates("#SampleID")
df2.head()
```

```python
#filter the data set to remove cats housed in shelters
simCats =  df2[(df2['SocializationStatus'] == 'House')]
```

```python
#Make a graphic of cat breeds in the data set
#First clean up the breed column, then calculate percentages of each breed
breed1 = simCats[['Breed']]
breed1.head()

#fill nan
breedNA = breed1.fillna(0)

# #replace messy values
breedNA.Breed.replace(['Unk0wn'],['Unknown'],inplace=True)

# #replace 0 values
breedNA.Breed.replace([0],['Unknown'],inplace=True)
breedNA.Breed.unique()

#calculate percentagees of categorical variables
breedNA['count'] = 1
breedNA2 = breedNA.groupby('Breed', sort=True).sum()/len(breedNA)
breed = breedNA2.reset_index()
breed.head()
```

```python
#Make a pie chart of cat breeds in the data set. 

#set up the figure
fig=plt.figure()
fig.set_size_inches(8,6)
sns.set(font='serif')

#Create the figure with two rows and three columns. Figsize will define the figure size
#create subplot panels, define colors
fig, (ax1) = plt.subplots(nrows=1, ncols=1)
colors = ["#a651cb","#64b546","#db5b2f","#4c7cdc","#b6b633","#6e62d3","#d99234","#5d6bab",
          "#a6a758","#e26fd0","#5ab880","#e53580","#3dbbb8","#ce3b46","#62a3da","#a2522e",
          "#b891e2","#497938","#b4398f","#846e29","#8a549e","#d89767","#d85481","#db7a76","#d687b6","#9b4863"]

#Create first pie chart with labels 
#To avoid labels being too cluttered, labels only shown if > 0.011 (you need to play
#with this number for each individual data set), or stored as blank strings '')
labels2 = [n if v > 0.011 else ''
            for n, v in zip(breed['Breed'], breed['count'])]
ax1.pie(breed['count'], startangle=45, colors=colors, labels = labels2) 
ax1.axis('equal')
fig.suptitle('Cat breeds in the data set', fontsize=15)

plt.show()
fig.savefig('cat_breeds.png')
```

```python
#Make a graphic of indoor and outdoor cats in the data set
#First clean up the IndoorOutdoor column, then calculate percentages of each 
inOut = simCats[['IndoorOutdoor']]
inOut.head()

#fill nan
inOutNA = inOut.fillna('Unknown')
inOutNA.IndoorOutdoor.unique()

# #replace messy values
inOutNA.IndoorOutdoor.replace(['IndoorOutdoor', 'Every day'],['Both', 'Unknown'],inplace=True)
inOutNA.IndoorOutdoor.unique()

#calculate percentagees of categorical variables
inOutNA['count'] = 1
inOutNA2 = inOutNA.groupby('IndoorOutdoor', sort=True).sum()/len(inOutNA)
IOut = inOutNA2.reset_index()
IOut.head()
```

```python
#Indoor/Outdoor plot
#set up the figure
fig=plt.figure()
fig.set_size_inches(8,6)
sns.set(font='serif')

#Create the figure with two rows and three columns. Figsize will define the figure size
#create subplot panels, define colors

fig, (ax1) = plt.subplots(nrows=1, ncols=1)
colors = ["#a651cb","#64b546","#db5b2f","#4c7cdc","#b6b633","#6e62d3","#d99234","#5d6bab",
          "#a6a758","#e26fd0","#5ab880","#e53580","#3dbbb8","#ce3b46","#62a3da","#a2522e",
          "#b891e2","#497938","#b4398f","#846e29","#8a549e","#d89767","#d85481","#db7a76","#d687b6","#9b4863"]

labels2 = IOut['IndoorOutdoor']
ax1.pie(IOut['count'], startangle=0, colors=colors, labels = labels2) #autopct='%1.1f%%' : add this in order to add % values to pies
ax1.axis('equal')

fig.suptitle('Distribution of Indoor to Outdoor Cats in the Data Set', fontsize=15)
plt.show()
fig.savefig('cat_inOut.png')
```

```python
#Make new column for feeding information:
simCats.Feeding.unique()
# Clean up breed column, calculate percentages
feeding = simCats[['Feeding']]
feeding.head()

#calculate percentagees of categorical variables
feeding['FeedingCount'] = 1
feeding.head()
feeding2 = feeding.groupby('Feeding', sort=True).sum()/len(feeding)
fed = feeding2.reset_index()
print fed
```

```python
#Ok, I did this same thing so many times I made a function

def categorical_percentages(data, categorical_data_title):
    '''
    Calculate percentages of categorical variables.
    data = df['count_column_title']
    categorical_data_title = df['categorical data title']
    '''
    data['count'] = 1
    data2 = data.groupby(categorical_data_title, sort=True).sum()/len(data)
    percentages = data2.reset_index()
    return percentages
```

```python
#Feeding and Indoor/Outdoor plots
#set up the figure
fig=plt.figure()
fig.set_size_inches(10,6)
sns.set(font='serif')

#Create the figure with two rows and three columns. Figsize will define the figure size
#create subplot panels, define colors

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
colors = ["#a651cb","#64b546","#db5b2f","#4c7cdc","#b6b633","#6e62d3","#d99234","#5d6bab",
          "#a6a758","#e26fd0","#5ab880","#e53580","#3dbbb8","#ce3b46","#62a3da","#a2522e",
          "#b891e2","#497938","#b4398f","#846e29","#8a549e","#d89767","#d85481","#db7a76","#d687b6","#9b4863"]

#Create first pie chart with labels (labels only shown if > 0.08, or stored as blank strings '')

labels1 = fed['Feeding']
ax1.pie(fed['FeedingCount'], startangle=180, colors=colors, labels = labels1) #autopct='%1.1f%%' : add this in order to add % values to pies
ax1.set_xlabel('Types of food')
ax1.axis('equal')

labels2 = IOut['IndoorOutdoor']
ax2.pie(IOut['count'], startangle=0, colors=colors, labels = labels2) #autopct='%1.1f%%' : add this in order to add % values to pies
ax2.set_xlabel('Indoor/outdoor')
ax2.axis('equal')

plt.tight_layout()
fig.suptitle('Attributes of all cats in data set', fontsize=15)
plt.show()
fig.savefig('cat_feeding_inOut.png', bbox_inches='tight') #
```

```python
#And since I was making the same plot a lot, I made a pie plotting function too. 

def piePlot(data, labels, title, figName='figure.png', startangle=0): 
    '''
    title = str
    figName = str
    labels and data must be the same length
    labels = df['columnNameLabels']
    data = df['columnNameData']
    startangle = int (default is 0)
    '''
    fig=plt.figure()
    fig.set_size_inches(8,6)
    sns.set(font='serif')

    fig, (ax1) = plt.subplots(nrows=1, ncols=1)
    colors = ["#a651cb","#64b546","#db5b2f","#4c7cdc","#b6b633","#6e62d3","#d99234","#5d6bab",
              "#a6a758","#e26fd0","#5ab880","#e53580","#3dbbb8","#ce3b46","#62a3da","#a2522e",
              "#b891e2","#497938","#b4398f","#846e29","#8a549e","#d89767","#d85481","#db7a76","#d687b6","#9b4863"]

    ax1.pie(data, startangle=startangle, colors=colors, labels = labels) #autopct='%1.1f%%' : add this in order to add % values to pies
    ax1.axis('equal')

    fig.suptitle(title, fontsize=15)
    fig.savefig(figName)
    
    return plt.show()
```

```python
#Calling the function
piePlot(fed['FeedingCount'], fed['Feeding'], 'Distribution of feeding types', 'cat_feeding.png', 180)
```
